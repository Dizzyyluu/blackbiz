```python
from bs4 import BeautifulSoup
import requests

def get_url(page):
    site = 'https://byblacks.com/directory/business/list-alpha/{}'
    url = site.format(page)
    return url

def get_record(card):
    name = card.find("div", class_="mt-ls-header").text.replace('\t', '').replace('\n','')
    location_elem = card.find("div", class_="mt-ls-fields address")
    if location_elem:
        location = location_elem.text.replace('\t', '').replace('\n','')
    else:
        location = None
    category =  card.find("span", class_="output").text.replace('\t', '').replace('\n','')
    telephone_elem = card.find("div", {"class": "mt-ls-field mfieldtype_coretelephone span12 lastFieldRow"})
    if telephone_elem:
        telephone = telephone_elem.find("span", class_="output").text
    else:
        telephone = None
    if location is not None:
        record = ('Company name:\n', name, "\n\n", 'Location:\n', location, "\n\n", 'Category:\n', category, "\n\n", 'Telephone:\n', telephone)
    else:
        record = ('Company name:\n', name, "\n\n", 'Category:\n', category, "\n\n", 'Telephone:\n', telephone)

    return record

def scrape_all_results():
    records = []
    
    page = 1
    while True:
        url = get_url(page)
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        cards = soup.find_all("div", class_="lsrow row-fluid")
        
        if not cards:
            break
            
        for card in cards:
            record = get_record(card)
            records.append(record)
            
        page += 1
    
    return records

if __name__ == '__main__':
    
    records = scrape_all_results()
    print('\n')
    print(f"Total records found: {len(records)}" + '\n')
    
    for record in records:
        print(*record, sep='')
        print('\n')
        print('*------------*')
```
